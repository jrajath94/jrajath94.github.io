```python
%matplotlib inline
import ipywidgets
import datetime as dt
import numpy as np
import pandas as pd
import numpy as np
import matplotlib
import bqplot
import matplotlib.pyplot as plt


```

```python
illinois_buildings = pd.read_csv('building_inventory.csv')
buildings = illinois_buildings[illinois_buildings['Year Acquired'] != 0]
buildings = buildings[buildings['Year Constructed'] != 0]
```

```python
build_agency = np.array(buildings['Agency Name'])

build_district = np.array(buildings['Congress Dist'])
```

```python
buildings_departments = buildings.groupby(['Congress Dist','Agency Name'])['Square Footage'].sum().reset_index()
```

```python
buildings_pivot = buildings_departments.pivot_table(index='Congress Dist', columns='Agency Name',values=['Square Footage'])
buildings_pivot
```

```python
buildings_values  = buildings_pivot.values
buildings_values
```

```python

col_sc = bqplot.ColorScale(scheme = 'Spectral',min=np.nanmin(buildings_pivot), 
                           max=np.nanmax(buildings_pivot))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')
x_ax = bqplot.Axis(scale = x_sc)
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical')


heat_map = bqplot.GridHeatMap(color = buildings_pivot,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column':x_sc },
                              interactions = {'click': 'select'}, # make interactive on click of each box
                              anchor_style = {'fill':'blue'}, # to make our selection blue
                              selected_style = {'opacity': 1.0}, # make 100% opaque if box is selected
                              unselected_style = {'opacity': 0.8}) # make a little see-through if not

# stir and combine into 1 figure
fig = bqplot.Figure(marks = [heat_map], 
                    axes = [c_ax, y_ax, x_ax])

fig
```

```python
buildings_year_square = buildings.groupby('Year Acquired')['Square Footage'].sum().reset_index()

build_ya= np.array(buildings_year_square['Year Acquired'].astype('float'))
build_sq = np.array(buildings_year_square['Square Footage'].astype('float'))
```

```python
x_scl = bqplot.LinearScale(min=1753,max=2019) # note: for dates on x-axis
y_scl = bqplot.LinearScale(min=500, max=5000000)
width_line = bqplot.LinearScale()
ax_xcl = bqplot.Axis(label='Date', scale=x_scl)
ax_ycl = bqplot.Axis(label='Duration in Sec', scale=y_scl, 
                    orientation='vertical', side='left')

duration_Lines = bqplot.Lines(x = build_ya,
                               y = build_sq, 
                              scales={'x':x_scl, 'y':y_scl,'width': width_line,}
                                      )


fig_dur = bqplot.Figure(marks = [duration_Lines], axes = [ax_xcl, ax_ycl])
fig_dur
```
